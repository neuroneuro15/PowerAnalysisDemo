# Interpreting Data: Considering P-Value and Statistical Power

## Beware the Variance

```python
distribution = np.random.uniform(low=0., high=1., size=100)
_ = plt.hist(distribution, bins=15)
```

## What does it take to be Significant?

```python
p = stats.binom_test(x=10, n=14, alternative='two-sided')
p
```

## Simulating Experiments

```python
n, universes, bias = 40, 100, .66
data = np.random.uniform(low=0., high=1., size=(40, 100))
fig, (ax1, ax2) = plt.subplots(nrows=2)
_ = ax1.imshow(data, imlim=[0, 1])
_ = ax2.imshow(data > bias, cmap='gray')
```

## Consider Your Power

```python
n, universes, bias = 40, 100, .55
data = np.random.uniform(low=0., high=1., size=(n, universes))
wins = (data < bias).sum(axis=0)
p_vals = np.array([stats.binom_test(win, n) for win in wins])
_ = plt.hist(p_vals)
plt.title(f'{np.sum(p_vals<.05)} of {len(p_vals)} Experiments < p=.05')
```
